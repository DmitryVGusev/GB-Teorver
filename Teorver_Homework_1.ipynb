### –ó–∞–¥–∞—á–∞ 3

–°–µ–º—å —á–µ–ª–æ–≤–µ–∫ —Ä–∞—Å—Å–∞–∂–∏–≤–∞—é—Ç—Å—è –Ω–∞—É–¥–∞—á—É –Ω–∞ —Å–∫–∞–º–µ–π–∫–µ. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞ –±—É–¥—É—Ç —Å–∏–¥–µ—Ç—å —Ä—è–¥–æ–º?

#### –†–µ—à–µ–Ω–∏–µ

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ø–µ—Ä–≤—ã–º —Å–µ–ª –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π —á–µ–ª–æ–≤–µ–∫. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —ç—Ç–æ–≥–æ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç $\frac{2}{7}$. –ü—Ä–∏ —ç—Ç–æ–º, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ø—Ä–∞–≤–µ–µ —Å–µ–ª –≤—Ç–æ—Ä–æ–π –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π —á–µ–ª–æ–≤–µ–∫ - $\frac{1}{6}$. –ü—Ä–æ—Ö–æ–¥—è—Å—å –ø–æ –≤—Å–µ–º –º–µ—Å—Ç–∞–º –Ω–∞ –ª–∞–≤–æ—á–∫–µ, –∫—Ä–æ–º–µ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ, –∏—Å–ø–æ–ª—å–∑—É—è —Ñ–æ—Ä–º—É–ª—É –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏, –Ω–∞—Ö–æ–¥–∏–º –æ–±—â—É—é –≤–µ—Ä–æ—è—Ç–æ–Ω–æ—Å—Ç—å —á—Ç–æ –Ω–∞ –ª–∞–≤–æ—á–∫–µ –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º —Å–∏–¥—è—Ç –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞:
$$ P =  (\frac{2}{7} \cdot \frac{1}{6}) \cdot (7-1) = \frac{2}{7}$$

### –ó–∞–¥–∞—á–∞ 4

–ò–∑ $60$ –≤–æ–ø—Ä–æ—Å–æ–≤, –≤—Ö–æ–¥—è—â–∏—Ö –≤ —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω—ã–µ –±–∏–ª–µ—Ç—ã, —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç $50$. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ä–µ–¥–∏ —Ç—Ä—ë—Ö –Ω–∞—É–≥–∞–¥ –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –≤–æ–ø—Ä–æ—Å–æ–≤ —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç: –∞) –≤—Å–µ? –±) –¥–≤–∞?

#### –†–µ—à–µ–Ω–∏–µ

–∞) –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ —á—Ç–æ, —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç –≤—Å–µ –±–∏–ª–µ—Ç—ã = $\frac{50}{60} \cdot \frac{49}{59} \cdot \frac{48}{58} = 0.572$

```python
(50/60) * (49/59) * (48/58)
```

–∞) –ü—Ä–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–æ–≥–æ —á—Ç–æ, —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç –¥–≤–∞ –±–∏–ª–µ—Ç–∞, –º–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç —É–∂–µ –≤—ã–Ω—É–ª –æ–¥–∏–Ω –±–∏–ª–µ—Ç –∫–æ—Ç–æ—Ä—ã–π –æ–Ω –Ω–µ –∑–Ω–∞–µ—Ç, –∞ –∑–Ω–∞—á–∏—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å = $\frac{50}{59} \cdot \frac{49}{58} = 0.715$

```python
(50/59) * (49/58)
```

### –ó–∞–¥–∞—á–∞ 5

–ë—Ä–æ—Å–∞–µ—Ç—Å—è –∏–≥—Ä–∞–ª—å–Ω–∞—è –∫–æ—Å—Ç—å. –ü—É—Å—Ç—å —Å–æ–±—ã—Ç–∏–µ $A$ - –ø–æ—è–≤–ª–µ–Ω–∏–µ —á—ë—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞, —Å–æ–±—ã—Ç–∏–µ $B$ - –ø–æ—è–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ –±–æ–ª—å—à–µ —Ç—Ä—ë—Ö. –Ø–≤–ª—è—é—Ç—Å—è –ª–∏ —ç—Ç–∏ —Å–æ–±—ã—Ç–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏?

#### –û—Ç–≤–µ—Ç
–≠—Ç–∏ —Å–æ–±—ã—Ç–∏—è —Å–æ–≤–º–µ—Å—Ç–Ω—ã–µ, –Ω–æ –Ω–µ –∑–∞–≤–∏—Å–∏–º—ã–µ.
–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –ø–æ—è–≤–ª–µ–Ω–∏—è –¥–≤—É—Ö –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö —Å–æ–±—ã—Ç–∏–π –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ: $$P(AB) = P(A) \cdot P(B)$$

–í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏—è  $A - \frac{1}2 \space (2,4,6)$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏—è $B - \frac{1}2 \space (4,5,6) $

–ò –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥–≤—É—Ö –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Å–æ–±—ã—Ç–∏–π 
$$P(AB) = P(A) \cdot P(B) = \frac{1}2 \cdot \frac{1}2 = \frac{1}{4}$$

### –ó–∞–¥–∞—á–∞ 6

–î–æ–ø—É—Å—Ç–∏–º, –∏–º–µ–µ—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä–∞—è –æ—á–µ–Ω—å —Ä–µ–¥–∫–∞—è –±–æ–ª–µ–∑–Ω—å (–ø–æ—Ä–∞–∂–∞–µ—Ç $0.1 \%$ –Ω–∞—Å–µ–ª–µ–Ω–∏—è). –í—ã –ø—Ä–∏—Ö–æ–¥–∏—Ç–µ –∫ –≤—Ä–∞—á—É, –≤–∞–º –¥–µ–ª–∞—é—Ç —Ç–µ—Å—Ç –Ω–∞ —ç—Ç—É –±–æ–ª–µ–∑–Ω—å, –∏ —Ç–µ—Å—Ç –æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º. –í—Ä–∞—á –≥–æ–≤–æ—Ä–∏—Ç –≤–∞–º, —á—Ç–æ —ç—Ç–æ—Ç —Ç–µ—Å—Ç –≤–µ—Ä–Ω–æ –≤—ã—è–≤–ª—è–µ—Ç $99 \%$ –±–æ–ª—å–Ω—ã—Ö —ç—Ç–æ–π –±–æ–ª–µ–∑–Ω—å—é –∏ –≤—Å–µ–≥–æ –ª–∏—à—å –≤ $1 \%$ —Å–ª—É—á–∞–µ–≤ –¥–∞—ë—Ç –ª–æ–∂–Ω—ã–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç.

_–í–æ–ø—Ä–æ—Å_: –≤–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å–Ω—ã –µ–π?

_–ü–æ–¥—Å–∫–∞–∑–∫–∞_: –≤–Ω–æ–≤—å –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞ —Å —Ä–∞—Å–∫—Ä—ã—Ç–∏–µ–º –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—è —Å –ø–æ–º–æ—â—å—é —Ñ–æ—Ä–º—É–ª—ã –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏.

#### –†–µ—à–µ–Ω–∏–µ

–ü—É—Å—Ç—å —Å–æ–±—ã—Ç–∏–µ $A$ –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ —Ç–µ—Å—Ç –¥–∞–ª –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –∏ –ø—É—Å—Ç—å —Å–æ–±—ã—Ç–∏–µ $B_2$ –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ —è –±–æ–ª–µ–Ω —ç—Ç–æ–π –±–æ–ª–µ–∑–Ω—å—é –∏ $B_1$ - —á—Ç–æ –Ω–µ –±–æ–ª–µ–Ω. –ò–∑ —É—Å–ª–æ–≤–∏–π —Å–ª–µ–¥—É–µ—Ç, —á—Ç–æ —Ç–µ—Å—Ç –º–æ–∂–µ—Ç –¥–∞—Ç—å 1% –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ª–æ–∂–Ω–æ-–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –¥–∞–∂–µ –µ—Å–ª–∏ —è –Ω–µ –±–æ–ª–µ–Ω, —Ç–æ –µ—Å—Ç—å $P(A|B_1) = \frac{1}{100}$.

–ü—Ä–∏–º–µ–Ω—è—è —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞, –ø–æ–ª—É—á–∏–º:
$$
P(B_2\:|\:A) = \frac{P(A\:|\:B_2) \cdot P(B_2)}{P(A)} = 
\frac{P(A\:|\:B_2) \cdot P(B_2)}{P(A\:|\:B_2) \cdot P(B_2) + P(A\:|\:B_1) \cdot P(B_1)}= \\
= \frac
    {\frac{99}{100} \cdot \frac{1}{1000}}
    {\frac{99}{100} \cdot \frac{1}{1000} + \frac{1}{100} \cdot \frac{99}{1000}}
= \frac
    {\frac{99}{10^5}}
    {\frac{99}{10^5} + \frac{999}{10^5}} 
= \frac{99}{1098} = 0.09
$$

```python
99/(99+999)
```
